## Q1. Computing Cross-Correlation

We provide here a standardization function

```python
def stand(x):
    return (x-np.mean(x))/np.std(x, ddof=0)
```

Write a function that takes two time series and uses the fourier method to calculate a cross-correlation function

```python
def ccor(ts1, ts2):
    return nfft.ifft(np.multiply(nfft.fft(ts1),nfft.fft(ts2)))
```

We plot the correlation function. Where does it maximize?

**Answer**: It maximizes at the place where the shift gives the maximum correlation.

```python
plt.plot(x1, ccor(stand(c51), stand(c16)))
```

## Q2. Dimensionality reduction with SAX

The idea behind SAX is to reduce dinensionality by capturing pieces of the curve and getting an average. So you are replacing the function by a set of steps. The next part is to encode the steps using some technique, here we will use binary strings. The [iSAX paper](https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=1&ved=0ahUKEwi2y4u9p__LAhUD9R4KHYslCuAQFggdMAA&url=http%3A%2F%2Fwww.cs.ucr.edu%2F~eamonn%2FiSAX.pdf&usg=AFQjCNEhDE8xHlXBVfSkLUBo4UsVRHOyTw&sig2=dgG-oNiR2MAYcqyUWzCs3Q), which you should skim, at the very least, is the source of this explanation below

![](https://dl.dropboxusercontent.com/u/75194/saxrepr.png)



Our interpolated curves have 1024 points. Bin them in chunks of 32, so we have $w=32$ chunks, and use a cardinality $a=4$ to get a SAX(T, 32, 4). Represent both light curves this way. For bonus points solve it for a general cardinality a (you'll have to get the gaussian division from the table above).

Remember you are doing this on standardized time series. You can assume that the series has been interpolated to a power of 2 length. Write a function to do this below:

```python
np.median(range(11))
```

```python
def representer(series, w, a=4): # 'a' will be 4
    # Assume input 'series' is already standardized
    Npoints = series.shape[0]
    assert(Npoints == 1024) # make sure it has the right number of points
    
    step = Npoints / 32 
    assert(step == int(step))# should be an integer
    
    cutoffs = [-np.inf, -0.67, 0, 0.67, np.inf]
    Ncutoffs = 5
    
    outputSeries = np.zeros(w,)
    
    for i in range(w): # compute for each bin
        minInd = i*step
        maxInd = (i + 1)*step
        medVal = np.median(series[minInd:maxInd])
        for j in range(1,Ncutoffs):
            if medVal < cutoffs[j]:
                outputSeries[i] = j-1
                break
                
    return outputSeries
```

We apply the function

```python
representer(stand(c51), 32)
```

```python
plt.plot(representer(stand(c51), 32))
```

```python
plt.plot(representer(stand(c16), 32))
```
